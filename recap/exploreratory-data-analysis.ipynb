```python
"""Answering Research Questions Based on The Mexico Real Estate Dataset"""

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv("data/mexico-real-estate-clean.csv")

#Question 1: Do Housing prices vary by state?

#to answer, we plot the mean prices grouped by state on Bar charts to visualize the variation.

mean_price_by_state = df.groupby("state")["price_usd"].mean().sort_values(ascending = False)
mean_price_by_state.plot(
    kind="bar",
    ylabel ="Price in Usd",
    xlabel = "state",
    title = "Mean price in USD per State"
);

```

```python
"""What is the price per square Metres"""

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv("data/mexico-real-estate-clean.csv")


#Calculating price per square Metres and adding that to our DataFrame df

df["price_per_m2"] = df["price_usd"]/df['area_m2']
(
    df
    .groupby("state")["price_per_m2"]
    .mean().sort_values(ascending = False)
    .plot(
    kind = "bar",
    xlabel ="state",
    ylabel="Mean Price per square metres",
    title = "Mean house price per square meters by state",
    )
);
```

```python
"""Is there a relationship between price and home sizes"""

#to answer the above question, we'll a scatterplot to visualize the corelation between the two datas

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv("data/mexico-real-estate-clean.csv")

#Drawing the scatterplot

plt.scatter(x = df["area_m2"], y = df["price_usd"])
plt.xlabel("Area in square meters")
plt.ylabel("Price (USD)")
plt.title("Area vs Price");
```

```python
import pandas as pd  


df = pd.read_csv("data/mexico-real-estate-clean.csv")



"""Determining The Correlation between area and price"""

#Correlation is a statistical measure that descirbes 
# the extent of a linear relationship between two variables

#using pandas we can use the function corr() to calculate the correlation between two columns
#ie for this case we can use the columns "area_m2" and "price_usd"

p_correlation =df["area_m2"].corr(df["price_usd"])

print("p_correlation:", p_correlation)
```
