```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a DataFrame with the data (you'll need to estimate the exact values from the bars)
data = {
    'Year': [2005, 2010, 2015, 2020],
    'H5N1': [5, 15, 20, 10],  # Estimated values
    'H5N2': [2, 8, 12, 5],
    'H5N6': [0, 3, 15, 8],
    'H5N8': [0, 5, 18, 12],
    'H7N1': [1, 2, 3, 1],
    'H7N3': [2, 4, 5, 3],
    'H7N7': [1, 3, 4, 2],
    'H7N9': [0, 5, 12, 8],
    'Multiple subtypes': [3, 8, 10, 6]
}

df = pd.DataFrame(data)
df.set_index('Year', inplace=True)

# Create stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Get unique colors for each subtype
colors = plt.cm.tab20(np.linspace(0, 1, len(df.columns)))

# Plot each subtype
bottom = np.zeros(len(df))
for i, col in enumerate(df.columns):
    ax.bar(df.index, df[col], bottom=bottom, label=col, color=colors[i])
    bottom += df[col]

# Customize the plot
ax.set_title('Temporal Description - Distribution of Publication Date for Different Avian Influenza Subtypes', pad=20)
ax.set_xlabel('Year of Publication')
ax.set_ylabel('Number of Article(s)')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
"""Answering Research Questions Based on The Mexico Real Estate Dataset"""

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv("data/mexico-real-estate-clean.csv")

#Question 1: Do Housing prices vary by state?

#to answer, we plot the mean prices grouped by state on Bar charts to visualize the variation.

mean_price_by_state = df.groupby("state")["price_usd"].mean().sort_values(ascending = False)
mean_price_by_state.plot(
    kind="bar",
    ylabel ="Price in Usd",
    xlabel = "state",
    title = "Mean price in USD per State"
);

```

```python
"""What is the price per square Metres"""

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv("data/mexico-real-estate-clean.csv")


#Calculating price per square Metres and adding that to our DataFrame df

df["price_per_m2"] = df["price_usd"]/df['area_m2']
(
    df
    .groupby("state")["price_per_m2"]
    .mean().sort_values(ascending = False)
    .plot(
    kind = "bar",
    xlabel ="state",
    ylabel="Mean Price per square metres",
    title = "Mean house price per square meters by state",
    )
);
```

```python
"""Is there a relationship between price and home sizes"""

#to answer the above question, we'll a scatterplot to visualize the corelation between the two datas

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv("data/mexico-real-estate-clean.csv")

#Drawing the scatterplot

plt.scatter(x = df["area_m2"], y = df["price_usd"])
plt.xlabel("Area in square meters")
plt.ylabel("Price (USD)")
plt.title("Area vs Price");
```

```python
"""Determining The Correlation between area and price"""

import pandas as pd  


df = pd.read_csv("data/mexico-real-estate-clean.csv")

#Correlation is a statistical measure that descirbes 
# the extent of a linear relationship between two variables

#using pandas we can use the function corr() to calculate the correlation between two columns
#ie for this case we can use the columns "area_m2" and "price_usd"

p_correlation =df["area_m2"].corr(df["price_usd"])

print("p_correlation:", p_correlation)
```

```python
"""Exploring the relationship Between Area and price in individual states, in this Case we use the MORELOS STATE"""

#To achieve this, we create a new dataframe that only consists of rows where state is Morelos

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/mexico-real-estate-clean.csv")

df_morelos = df[df["state"] == "Morelos"]

#Using the df_morelos Dataframe, a subset of the df Dataframe, 
#we'll visualize a scatter plot do dtermine the relationship bnetween area and price in the Morelos state

plt.scatter(x=df_morelos["area_m2"], y=df_morelos["price_usd"])
plt.xlabel("Area (m²)")
plt.ylabel("Price (USD)")
plt.title("Morelos: Area vs Price")
plt.show()
```
